Let's manually extract each rows. since some rows contains two columns, we need to seperate them. So I have used x_threshold to seperate each columns within a row, and add it as a separete row.

```python
def extract_row_lines(ocr_output, y_threshold=40, x_threshold=400):
    """
    Extracts row lines from PaddleOCR output based on bounding box positions and thresholds.

    Args:
        ocr_output (list): List of OCR outputs, each containing a bounding box and text.
        y_threshold (int): Maximum difference in the y-coordinate to group bounding boxes into a row.
        x_threshold (int): Maximum difference in the x-coordinate to group bounding boxes into different sub-rows.

    Returns:
        list: A list of rows, where each row is a list of text elements found in that row.
    """
    rows = []
    current_row = []
    row_center = 0
    last_x = None

    for box in ocr_output:
        # Extract the top-left and bottom-left y-coordinates of the bounding box
        top_left_x = box[0][0][0]
        top_left_y = box[0][0][1]
        bottom_left_y = box[0][3][1]
        cell_center_y = (top_left_y + bottom_left_y) / 2

        # Initialize the row center if it's the first element
        if row_center == 0:
            row_center = cell_center_y
        print(f"Text: {box[1][0]}")
        print(f"Y_difference: {abs(row_center - cell_center_y)}")

        # Check if the y-difference exceeds the y_threshold, meaning a new row should start
        if abs(row_center - cell_center_y) > y_threshold:
            # Append the current row to rows
            if current_row:
                rows.append(current_row)
            # Reset row tracking
            row_center = cell_center_y
            current_row = [box[1][0]]  # Start a new row with the current text
            last_x = top_left_x  # Reset the x position
        else:
            # If still within the same y-level, check x_threshold
            if last_x is not None and abs(top_left_x - last_x) > x_threshold:
                # If x difference is too large, treat it as a new sub-row
                rows.append(current_row)
                current_row = [box[1][0]]
            else:
                # Otherwise, continue adding to the current row
                current_row.append(box[1][0])

            last_x = top_left_x  # Update the last x position

    # Append the last row if it has content
    if current_row:
        rows.append(current_row)

    return rows
```

```python
from fuzzywuzzy import fuzz
import json

def extract_key_value_pairs_fuzzy(lines, keys, threshold=80):
    """
    Extract key-value pairs from lines of text using fuzzy matching for keys (fuzz.partial_ratio).

    Args:
        lines (list): List of strings representing the text lines.
        keys (list): List of keys to search for in the text.
        threshold (int): Minimum fuzzy match score to consider a match valid (default: 80).

    Returns:
        dict: Extracted key-value pairs.
    """
    key_value_pairs = {}

    for line in lines:
        # Iterate over each key to find the best fuzzy match
        for key in keys:
            match_ratio = fuzz.partial_ratio(key.lower(), line.lower())

            # If match ratio is above the threshold, consider it a match
            if match_ratio >= threshold:
                # Extract the value before the key and assign the rest of the line as the value
                key_index = line.lower().find(key.lower())

                # If the key is found, split the line into the part before and after the key
                if key_index != -1:
                    value_before_key = line[:key_index].strip()
                    value_after_key = line[key_index + len(key):].strip()

                    # Combine the extracted value and assign to the matched key
                    value = f"{value_before_key} {value_after_key}".strip()
                    key_value_pairs[key] = value

                # Remove the matched key from the keys list
                keys.remove(key)
                break  # Exit loop after finding a match for this line

    return key_value_pairs
```

In the implemented logic for extracting key-value pairs from OCR text data, two primary techniques are used. row segmentation and fuzzy matching.

Row Segmentation Logic: The extract_row_lines function processes bounding box data from PaddleOCR and groups text into rows based on y-coordinates. A y_threshold helps separate rows vertically, while an x_threshold handles gaps between words in the same row.

Fuzzy Matching for Key Extraction: The extract_key_value_pairs_fuzzy function uses fuzzy matching to identify keys in each row. It finds partial matches for predefined keys and extracts corresponding values, making it robust to minor variations in OCR outputs.

```python
def extract_key_values_paddle(ocr_output, keys, value_threshold=3, fuzzy_threshold=80, x_threshold=100):

  for i, (coords, (token, confidence)) in enumerate(ocr_output):
      for key in keys:
          if fuzz.partial_ratio(token.lower(), key.lower()) >= fuzzy_threshold:
              print(f"Matching Token: '{token.lower()}' with Key: '{key.lower()}'")

              # Extract potential value, ensuring to remove leading/trailing spaces and colons
              potential_value = token[len(key):].strip().lstrip(':').strip()

              if len(potential_value) >= value_threshold:
                  extracted_data[key] = potential_value
                  keys.remove(key)
                  break  # Stop checking other keys for this token

              for j in range(i + 1, len(ocr_output)):
                  next_coords, (next_token, next_confidence) = ocr_output[j]
                  current_center_y = (coords[1][1] + coords[2][1]) / 2
                  next_center_y = (next_coords[1][1] + next_coords[2][1]) / 2

                  if abs(current_center_y - next_center_y) < (coords[2][1] - coords[1][1]):
                      print(f"Next Token Found: '{next_token}' at Coordinates: {next_coords}")

                      if next_coords[0][0] >= coords[1][0] and abs(next_coords[0][0] - coords[1][0]) < x_threshold:
                          extracted_data[key] = next_token
                          keys.remove(key)
                          break
                      elif abs(current_center_y - next_center_y) < 2 * (coords[2][1] - coords[1][1]):
                          if abs(next_coords[0][0] - coords[0][0]) < (x_threshold * 0.5):
                              extracted_data[key] = next_token
                              keys.remove(key)
                              break

  return extracted_data
```

```python
import spacy

# Load the spaCy model for NER
nlp = spacy.load("en_core_web_sm")

# Step 1: Define the mapping between keys and expected spaCy entity labels
key_entity_mapping = {
    'Account Name': 'ORG',
    'Account Number': 'CARDINAL',
    'Account Type': 'PRODUCT',          # Could be types like 'Savings', 'Checking', better approximated by 'ORG'
    'Statement Period': 'DATE',
    'Balance': 'MONEY',
    'Total money in': 'MONEY',
    'Total money out': 'MONEY',
}
```

```python
# Step 2: Function to check if the value matches the expected entity type
def validate_value(key, value):
    if key not in key_entity_mapping:
        return True  # Key not in the mapping

    expected_entity = key_entity_mapping[key]

    # Run the spaCy NER model on the value
    doc = nlp(value)
    print(doc.ents)
    # Check if any entity in the doc matches the expected entity type
    for ent in doc.ents:
        print(ent.label_ , expected_entity)
        if ent.label_ == expected_entity:
            return True  # Value matches the expected entity type

    return False  # No match found
```

### Why NER Doesn’t Work Well on Isolated Words:
NER models are trained on sentences or longer text segments where the context provides essential information. Here's why they struggle with single words or very short phrases:

*   **Context Dependenc**y: NER models rely heavily on the context of a word to understand what entity it represents. For example, the word "Amazon" could refer to :


> A company (ORG) in the sentence "Amazon announced a new product."


> A geographical region (GPE) in the sentence "I traveled to the Amazon." Without the surrounding words, the model has no clue about which entity "Amazon" refers to.


*   **Ambiguity:** Many words have multiple meanings (like "Apple" for the company or the fruit). The model resolves this ambiguity by looking at the surrounding text. If you give it only one word, it has no information to resolve the ambiguity.


*   **Lack of Features:** Machine learning models, including NER models, are trained to look at features from the whole sentence. Features like neighboring words, word order, syntactic roles, and sentence patterns all contribute to the model’s understanding. With just a single word, many of these features are missing, making it difficult for the model to make an accurate prediction.

*   **Training Data:** During training, NER models are usually fed full sentences or paragraphs. As a result, they learn to depend on the surrounding context to make decisions about entity types. They aren’t typically trained to recognize named entities in isolation.













```python
import re

def validate_key_value(key, value):
    patterns = {
        'Account Name': r'^[A-Za-z0-9.,\s]+$',  # Alphanumeric, periods, commas, and spaces
        'Account Number': r'^\d{1,}$',           # One or more digits
        'Account Type': r'^^[A-Za-z]+$',  # Specific account types
        'Statement Period': r'^\d{2}/\d{2}/\d{4} to \d{2}/\d{2}/\d{4}$',  # MM/DD/YYYY to MM/DD/YYYY
        'Balance on June 1': r'^\$\d{1,3}(,\d{3})*(\.\d{2})?$',  # $X,XXX.XX
        'Total money in': r'^\$\d{1,3}(,\d{3})*(\.\d{2})?$',  # $X,XXX.XX
        'Total money out': r'^\$\s?\d{1,3}(,\d{3})*(\.\d{2})?$', # $X,XXX.XX
        'Balance on June 30': r'^\$\d{1,3}(,\d{3})*(\.\d{2})?$'  # $X,XXX.XX
    }

    pattern = patterns.get(key)
    if pattern:
        # Strip any leading ':' or whitespace from the value
        clean_value = value.lstrip(':').strip()
        if re.match(pattern, clean_value):
            return True
    return False
```
